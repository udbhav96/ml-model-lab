```python
import numpy as np # linear algebra
import pandas as pd
from sklearn.impute import KNNImputer
from sklearn import metrics

from sklearn.model_selection import train_test_split


from sklearn.linear_model import LinearRegression

import matplotlib.pyplot as plt
```

```python
AD_dataset = pd.read_csv('/home/udbhav-purwar/Documents/Project/ML_Models/Linear Regression/Admission_Predict.csv')
```

```python
AD_dataset.head()
```

```python
AD_dataset.info()
```

```python
AD_dataset.isnull().sum()
```

```python
X = AD_dataset.drop(['Serial No.' , 'Chance of Admit '],axis=1)
Y = AD_dataset['Chance of Admit ']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
lin_reg_model = LinearRegression()
```

```python
lin_reg_model.fit(X_train , Y_train)
```

```python
training_data_prediction = lin_reg_model.predict(X_train)
```

```python
error_score = metrics.r2_score(Y_train, training_data_prediction )
print("R square error:", error_score)
```

```python
plt.scatter(Y_train , training_data_prediction)
plt.xlabel("Actual")
plt.ylabel("Pridicted ")
plt.title("Actual  VS Pridicted")
plt.show()
```

```python
test_data_prediction = lin_reg_model.predict(X_test)
```

```python
error_score = metrics.r2_score(Y_test, test_data_prediction )
print("R square error:", error_score)
```

```python
plt.scatter(Y_test , test_data_prediction)
plt.xlabel("Actual ")
plt.ylabel("Pridicted ")
plt.title("Actual  VS Pridicted ")
plt.show()
```
